```python
import pandas as pd
import numpy as np
import gc
```

```python
dfgames = pd.read_csv('games.csv')
dfgames.head()
```

```python
dfgames.info()
```

```python
for col in dfgames.columns:
    if dfgames[col].dtype == 'int64':
        dfgames[col] = dfgames[col].astype('int32')
    gc.collect()
```

```python
dfgames.info()
```

```python
dfpffscout = pd.read_csv('pffScoutingData.csv')
dfpffscout.head()
```

```python
dfpffscout.info()
```

```python
for col in dfpffscout.columns:
    if dfpffscout[col].dtype == 'int64':
        dfpffscout[col] = dfpffscout[col].astype('int32')
    if dfpffscout[col].dtype == 'float64':
        dfpffscout[col] = dfpffscout[col].astype('float32')
    gc.collect() 
```

```python
dfpffscout.info()
```

```python
dfgames_pffscout = dfgames.merge(dfpffscout,on='gameId')
dfgames_pffscout.head()
```

```python
dfgames_pffscout.info()
```

```python
dfplayers = pd.read_csv('players.csv')
dfplayers.head()
```

```python
dfplayers.info()
```

```python
dfplayers.info()
```

```python
for col in dfplayers.columns:
    if dfplayers[col].dtype == 'int64':
        dfplayers[col] = dfplayers[col].astype('int32')
    if dfplayers[col].dtype == 'float64':
        dfplayers[col] = dfplayers[col].astype('float32')
```

```python
dfplayers.info()
```

```python
dfgames_pffscout_players = dfgames_pffscout.merge(dfplayers,on='nflId')
dfgames_pffscout_players.head()
```

```python
dfgames_pffscout_players.info()
```

```python
dfplays = pd.read_csv('plays.csv')
dfplays.head()
```

```python
dfplays.info()
```

```python
for col in dfplays.columns:
    if dfplays[col].dtype == 'int64':
        dfplays[col] = dfplays[col].astype('int32')
    if dfplays[col].dtype == 'float64':
        dfplays[col] = dfplays[col].astype('float32')
    gc.collect()
```

```python
dfgames_pffscout_players_plays = dfgames_pffscout_players.merge(dfplays,on=['gameId','playId'])
dfgames_pffscout_players_plays.head()
```

```python
dfweek = pd.read_feather('optimized_nfl_new.feather')
dfweek.head()
```

```python
dffinal = dfgames_pffscout_players_plays.merge(dfweek,on=['gameId','playId','nflId'])
dffinal.head()
```
